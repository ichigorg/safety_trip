```python
def time_converter(second_col):
    time_period = []
    for time in second_col:
        if len(time_period) == 0:
            time_period.append(time)
            continue
        time_period.append(time_period[len(time_period)-1] + time)
    return time_period
```

```python
def change_converter (measurement):
    change = []
    for i in range(measurement.shape[0]):
        if len(change) == 0:
            change.append(math.sqrt((measurement.iloc[i, 0])**2 + (measurement.iloc[i, 1])**2 + (measurement.iloc[i, 2])**2))
            continue
        change.append(math.sqrt(((measurement.iloc[i, 0]) - (measurement.iloc[i-1, 0]))**2 + ((measurement.iloc[i, 1]) - (measurement.iloc[i-1, 1]))**2 + ((measurement.iloc[i, 2]) - (measurement.iloc[i-1, 2]))**2))
    return change
```

```python
class SafetyDecisionTree(tree.DecisionTreeClassifier):
    def __init__(self, criterion = 'entropy', splitter = 'best', max_depth = 6, min_samples_split = 4, min_samples_leaf = 7, max_features = 'auto'):
        #{'splitter': 'best', 'min_samples_leaf': 7, 'min_samples_split': 4, 'criterion': 'entropy', 'max_features': 'auto', 'max_depth': 6}
      self.model = tree.DecisionTreeClassifier(criterion = criterion, splitter = splitter, max_depth = max_depth, min_samples_split = min_samples_split, min_samples_leaf = min_samples_leaf, max_features= max_features)
    
    def fit(self, safety_features, safety_labels):
        self.__features = safety_features
        self.__labels = safety_labels.loc[:,'label']
        #self.__labels = pd.DataFrame([[1],[0],[1]], columns=['label'])
        self.__bookingID = safety_labels.loc[:,'bookingID']
        #self.__bookingID = pd.DataFrame([[1056561954943], [1202590843006],[996432412742]], columns = ['bookingID']).loc[:,'bookingID']
        self.__processed_features = self.__features_preprocessing()
        self.model.fit(self.__processed_features, self.__labels)
        
    def predict(self, safety_features, bookinID_dataframe):
        self.__features = safety_features
        self.__bookingID = bookinID_dataframe
        #self.__bookingID = pd.DataFrame([[712964571270], [1520418422978]], columns = ['bookingID']).loc[:,'bookingID']
        print(self.__bookingID.shape)
        self.__processed_features = self.__features_preprocessing()
        return self.model.predict(self.__processed_features)
    
    def __time_converter(self, bookingID):
        second_col = self.__features.loc[self.__features['bookingID'] == bookingID,'second']
        time_period = []
        for time in second_col:
            if len(time_period) == 0:
                time_period.append(time)
                continue
            time_period.append(time_period[len(time_period)-1] + time)
        time_period = pd.DataFrame(time_period, columns = ['time_period'])
        return time_period

    def __change_converter (self, bookingID, measurement_attr):
        if measurement_attr == 'accel':
            measurement = self.__features.loc[self.__features['bookingID'] == bookingID, ['acceleration_x','acceleration_y','acceleration_z']]
        elif measurement_attr == 'gyro':
            measurement = self.__features.loc[self.__features['bookingID'] == bookingID, ['gyro_x','gyro_y','gyro_z']]
        
        change = []
        for i in range(measurement.shape[0]):
            if len(change) == 0:
                change.append(math.sqrt((measurement.iloc[i, 0])**2 + (measurement.iloc[i, 1])**2 + (measurement.iloc[i, 2])**2))
                continue
            change.append(math.sqrt(((measurement.iloc[i, 0]) - (measurement.iloc[i-1, 0]))**2 + ((measurement.iloc[i, 1]) - (measurement.iloc[i-1, 1]))**2 + ((measurement.iloc[i, 2]) - (measurement.iloc[i-1, 2]))**2))
        name = measurement_attr + '_change'
        change = pd.DataFrame(change, columns = [name])
        return change
    
    def __magnitude_converter(self, bookingID, measurement_attr):
        if measurement_attr == 'accel':
            measurement = self.__features.loc[self.__features['bookingID'] == bookingID, ['acceleration_x','acceleration_y','acceleration_z']]
        elif measurement_attr == 'gyro':
            measurement = self.__features.loc[self.__features['bookingID'] == bookingID, ['gyro_x','gyro_y','gyro_z']]
        
        magnitude = []
        for i in range(measurement.shape[0]):
            magnitude.append(math.sqrt((measurement.iloc[i, 0])**2 + (measurement.iloc[i, 1])**2 + (measurement.iloc[i, 2])**2))
        name = measurement_attr + '_magnitude'
        magnitude = pd.DataFrame(magnitude, columns = [name])
        return magnitude
    
    def __change_reg(self, time_period, bookingID, measurement_attr):
        value = self.__change_converter(bookingID, measurement_attr)
        linreg = linear_model.LinearRegression()
        linreg.fit(time_period, value)
        return linreg
    
    def __magnitude_reg(self, time_period, bookingID, measurement_attr):
        value = self.__change_converter(bookingID, measurement_attr)
        linreg = linear_model.LinearRegression()
        linreg.fit(time_period, value)
        return linreg
    
    def __gps_cluster(self, bookingID):
        gps_cluster = cluster.KMeans(n_clusters=1)
        gps_cluster.fit_predict(features_df.loc[features_df['bookingID'] == bookingID, ['Accuracy','Bearing','Speed']])
        return gps_cluster
    
    def __features_preprocessing(self):
        dataframe = []
        count = 0
        for b_id in self.__bookingID:
            row = []
            time_period = self.__time_converter(b_id)
            accel_magnitude = self.__magnitude_reg(time_period, b_id, 'accel')
            accel_change = self.__change_reg(time_period, b_id, 'accel')
            gyro_change = self.__change_reg(time_period, b_id, 'gyro')
            gps_cluster = self.__gps_cluster(b_id)
            for value in gps_cluster.cluster_centers_[0]:
                row.append(value)
            row.append(accel_magnitude.intercept_[0])
            row.append(accel_magnitude.coef_[0,0])
            row.append(accel_change.intercept_[0])
            row.append(accel_change.coef_[0,0])
            row.append(gyro_change.intercept_[0])
            row.append(gyro_change.coef_[0,0])
            dataframe.append(row)
            count = count + 1
            print('\r'),
            print('Progress: {0:.2f}%'.format(count*100.00/self.__bookingID.shape[0])),
        dataframe = pd.DataFrame(dataframe, columns =['Accuracy_center', 'Bearing_center', 'Speed_center', 'min_accel', 'accel_rate','min_accel_change','accel_change_rate', 'min_gyro_change', 'gyro_change_rate'])
        return dataframe
     
```

```python
kfold = model_selection.KFold(n_splits=10, shuffle=True)
cv_acc = []
cv_error = []
for train, test in kfold.split(labels_df):
    cv_model = SafetyDecisionTree()
    cv_model.fit(features_df,labels_df.loc[train])
    pred = cv_model.predict(features_df, labels_df.loc[test].loc[:, 'bookingID'])
    scores = accuracy_score(labels_df.loc[test].loc[:, 'label'], pred)
    print("Accuracy score: %.2f%%" % (scores*100))
    cv_acc.append(scores * 100)
    cv_error.append(100-scores * 100)
print("Cross Validation Accuracy %.2f%% (+/- %.2f%%)" % (np.mean(cv_acc),np.std(cv_acc)))
print("Cross Validation Error %.2f%% (+/- %.2f%%)" % (np.mean(cv_error),np.std(cv_error)))
```
